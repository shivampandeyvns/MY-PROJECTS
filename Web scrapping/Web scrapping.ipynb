```python
html_text=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Machine+Learning&txtLocation=').text
print('Put some skills you are not familiar with (comma separated)')
unfamiliar_skill=input('>').lower().split(',')
print(f"\nfiltering out: {unfamiliar_skill}\n\n")
company_name_list=[]
skills_list=[]
more_info_list=[]
def find_jobs():
    soup=BeautifulSoup(html_text,'lxml')
    jobs=soup.find_all('li',class_ ='clearfix job-bx wht-shd-bx')
    for job in jobs:
        published_date=job.find('span',class_='sim-posted').text
        if 'few' in published_date:
            company_name=job.find('h3',class_ ="joblist-comp-name").text.strip()
            skills=job.find('span',class_="srp-skills").text.strip().replace(' ','')
            more_info=job.header.h2.a['href']
            check=True
            for skill in unfamiliar_skill:
                if skill in skills.lower():
                    check=False
            if(check==True):
                company_name_list.append(company_name)
                skills_list.append(skills)
                more_info_list.append(more_info)
                
    with open('job_list.txt','w') as f:
        print('Writing to the file...\n')
        for i in range(len(company_name_list)):
            f.write(f'Company name: {company_name_list[i]}\n')
            f.write(f'Skills: {skills_list[i]}\n')
            f.write(f'More info: {more_info_list[i]}\n')
            f.write('\n\n')
        print('Writing to the file completed\n')
if __name__=='__main__':
    while True:
        find_jobs()
        time_wait=10
        print(f'Waiting {time_wait} min....')
        time.sleep(time_wait*60)
```
